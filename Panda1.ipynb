# Pandas : Filter Data part2

```python
import pandas as pd
```

```python
df=pd.read_csv('master.csv')
df
```

```python
# I want data of after 2010 or 2010
mask1=df.year>=2010
df.loc[mask1]
```

```python
# I want data of 1989 to 2000
mask2=df.year.between(1989,2000)
df.loc[mask2]    # 1990 and 2000 both will include in result
```

```python
# i want only 2000's data and 2010's data
mask3=df.year.isin([2000,2010])
df.loc[mask3]
```

```python
f_df2=df.loc[mask3]
f_df2.year.unique()
```

```python
# Now I want whole year's data except 2000 and 2010
mask4=df.year.isin([2000,2010])
f_df1=df.loc[~mask4]
```

```python
f_df1.year.unique()
```

```python
# I want to compare suicide cases of male and female between 2010 and 2014 of switzerland and age shold between 15 to 24 and i want only country, age, sex, and suicide_no
```

```python
m1=df.year.between(2010,2014)
m2=df.sex=='male'
m3=df.sex=='female'
m4=df.country=='Switzerland'
m5=df.age=='15-24 years'
f_df3=df.loc[m1 & m2 & m4 & m5,['country','year','age','sex','suicides_no']]
f_df3
```

```python
f_df3.sum().suicides_no
```

```python
f_df4=df.loc[m1 & m3 & m4 & m5,['country','year','age','sex','suicides_no']]
f_df4
```

```python
f_df4.sum().suicides_no
```
